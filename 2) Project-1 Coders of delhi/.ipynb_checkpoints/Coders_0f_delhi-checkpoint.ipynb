```python
import json
def print_details(data):
    print("Users and Their Connections:")
    for user in data["users"]:
        print(f"{user["name"]} (ID: {user["id"]}) - Friends : {user["friends"]} - Liked Pages:{user["liked_pages"]}")
    print("\n\nPages")    
    for page in data["pages"]:  
        print(f"{page["id"]}: {page["name"]}")
        
def load_data(file):
    with open(file,"r") as f:
        data=json.load(f)
    return data
def write_data(data,file_name):
    with open(file_name,"w") as f:
        json.dump(data,f,indent=4)

```

```python
def clean_data(data):
    #remove empty name user and   #remove inactive user i.e user that have no friends and liked pages
     users=[user for user in data["users"] if user["name"].strip() and user["friends"] and user["liked_pages"]]
     #remove duplicate friend entry
     for user in users:
        user["friends"]=list(set(user["friends"]))
     
     unique_pages = {}
     for page in data["pages"]:
        unique_pages[page["id"]] = page
     data["pages"] = list(unique_pages.values())
    
     
     data["users"]=users
     
    
     
     return data
     
    
    
    
  
    
    
```

```python
data=load_data("data2.json")
clean_data=clean_data(data)
write_data(clean_data,"clean_data.json")
```
