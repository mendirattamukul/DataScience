```python
import matplotlib.pyplot as plt
import numpy as np
na=np.array([i for i in range(10)])
plt.subplot(1,2,1)
plt.xlabel("Square Root")
plt.ylabel("Number")

plt.plot(na,np.sqrt(na))
plt.subplot(1,2,2)
plt.plot(na,np.square(na))
plt.xlabel("Square")
plt.ylabel("Number")
plt.tight_layout()
```

```python
# 2by 2 grid layout
import matplotlib.pyplot as plt
import numpy as np
na=np.array([i for i in range(10)])
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.ylabel("Square Root")
plt.xlabel("Number")
plt.plot(na,np.sqrt(na))

plt.subplot(2,2,2)
plt.plot(na,np.square(na))
plt.ylabel("Square")
plt.xlabel("Number")


plt.subplot(2,2,3)
plt.plot(na,na/2)
plt.ylabel("Divid By 2")
plt.xlabel("Number")


plt.subplot(2,2,4)
plt.plot(na,na*2)
plt.ylabel("Multiply By 2")
plt.xlabel("Number")
plt.tight_layout()
```

```python
#subplots for clean code
import matplotlib.pyplot as plt
import numpy as np
na=np.array([i for i in range(10)])
fig,axs=plt.subplots(1,2,figsize=(10,5))
axs[0].plot(na,np.square(na))
axs[1].plot(na,np.sqrt(na))
axs[0].set_title("Square Representation Of Number")
axs[1].set_title("Square Root Representation Of Number")
fig.suptitle("Mathematics Representation Of Numbers")
fig.tight_layout()
fig.subplots_adjust(top=0.55)
fig.savefig("my_plots.png")
```

```python
#lopping over subplots
import matplotlib.pyplot as plt
import numpy as np
na=np.array([i for i in range(10)])
ys=np.array([[np.square(na),np.sqrt(na)],
    [na*2,na/2]])
titles=np.array([["Square","Square Root"],
               ["Double","Divide By 2"]])
fig,axs=plt.subplots(2,2,figsize=(8,5))

for i in range(2):
    for j in range(2):
        axs[i,j].plot(na,ys[i,j])
        axs[i,j].set_title(titles[i,j])
fig.tight_layout()
fig.suptitle("Subplots With Loop")
fig.subplots_adjust(top=0.75)

```
