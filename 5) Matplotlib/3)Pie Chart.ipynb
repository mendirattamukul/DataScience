```python
import matplotlib.pyplot as plt
months=["Jan-Mar","Apr-Jun","Jul-Sep","Oct-Dec"]
sales=[20,60,10,60]
colors=["orange","blue","red","#1BE393"]
explode=[0,0,0,0.1]
plt .pie(sales,
         labels=months,
         explode=explode,
         colors=colors,
        shadow=True,
        startangle=90,
        wedgeprops={"ec":"red"},
        autopct="%1.1f%%")
plt.title("Sale's record for last year")
```
