```python
import seaborn as sns
import matplotlib.pyplot as plt
print(sns.get_dataset_names())
```

```python
data=sns.load_dataset("tips")
```

```python
sns.lineplot(x="total_bill",y="tip",data=data)
plt.title("bill vs tips")
```

```python
sns.scatterplot(x="total_bill",y="tip",data=data,hue="time")
```

```python
sns.barplot(x="total_bill",y="tip",data=data,hue="time")
```

```python
sns.boxplot(x="total_bill",y="tip",data=data,hue="time")
```

```python
import pandas as pd
df=sns.load_dataset("flights")
pivot=df.pivot(index="month",columns="year",values="passengers")
sns.heatmap(pivot,annot=True,fmt="d",cmap="YlGnBu")
```
