```python
import matplotlib.pyplot as plt 
import numpy as np
ages_of_people_watching_my_youtube_videos=[10,15,16,13,14,12,11,
                                          55,54,53,52,51,56,
                                          25,26,36,34,42,44 
                                          ]
bins=[10,20,30,70]
plt.hist(ages_of_people_watching_my_youtube_videos,bins=10,edgecolor="black")
plt.xlabel("Ages")
plt.ylabel("frequency")
plt.axvline(np.mean(ages_of_people_watching_my_youtube_videos),color="red",linestyle="--",linewidth=2,label="Average")
plt.grid(True,linestyle="--",alpha=0.4)
plt.title("people watching me on youtube")
```
