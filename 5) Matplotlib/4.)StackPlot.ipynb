```python
import matplotlib.pyplot as plt
days=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
study=[1,2,1,3,4,1,0]
playing=[2,1,2,3,1,5,6]
sleeping=[8,10,8,7,6,10,12]
labels=["Study","Playing","Sleeping"]
colors=["green","orange","black"]
plt.grid(True)
plt.figure(figsize=(10,6))
plt.title("Weekly Schedule Plot")
plt.stackplot(days,study,playing,sleeping,labels=labels,colors=colors,alpha=0.4)
plt.legend(loc='upper left')
plt.xlabel("Days")
plt.ylabel("Activities")
plt.show()


```
