```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
 
plt.plot(years, runs)
plt.xlabel("Year")
plt.ylabel("Run")
plt.title("Sachin Tendulkar's Yearly Runs")
plt.show()
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
plt.plot(years,kohli,label="Kohli Run's")
plt.plot(years,sehwag,label="Sehwag Run's")
plt.xlabel("Year")
plt.ylabel("Run")
plt.title("Performance Comparision")
plt.legend()
plt.show()
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
plt.plot(years,kohli)
plt.plot(years,sehwag)
plt.xlabel("Year")
plt.ylabel("Run")
plt.title("Performance Comparision")
plt.legend(["kohli","sehwag"])#not recomended as it prone to mismatch
plt.show()
```

```python
#using format strings
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
plt.plot(years,kohli,"ro--",label="Kohli Run's")
plt.plot(years,sehwag,'g^:',label="Sehwag Run's")
plt.xlabel("Year")
plt.ylabel("Run")
plt.title("Performance Comparision")
plt.legend()
plt.show()
```

```python
#color and line style as argument
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
plt.plot(years,kohli,color="orange",linestyle="--",label="Kohli Run's")
plt.plot(years,sehwag,color="blue",linestyle=":",label="Sehwag Run's")
plt.xlabel("Year")
plt.ylabel("Run")
plt.title("Performance Comparision")
plt.legend()
plt.show()
```

```python
#line width and tweaks
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
plt.plot(years,kohli,color="orange",linestyle="--",linewidth=3,label="Kohli Run's")
plt.plot(years,sehwag,color="blue",linestyle=":",linewidth=2,label="Sehwag Run's")
plt.xlabel("Year")
plt.ylabel("Run")
plt.title("Performance Comparision")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.style.use("seaborn-v0_8")
plt.show()
```
