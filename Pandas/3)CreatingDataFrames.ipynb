```python
#From List of list
import pandas as pd
data=[
    ["Neha",27],
    ["Mukul",28]
]
df=pd.DataFrame(data,columns=["Name","Age"])
print(df)
```

```python
#From Dictionay Of Key As Column Name And list As Value
data2={"Name":["Mukul","Neha"],"Age":[28,27]}
df2=pd.DataFrame(data2)
print(df2)
```

```python
#from numpy arrays
import numpy as np
data3=np.array([[1,2],[3,4]])
df3=pd.DataFrame(data3,columns=["A","B"])
print(df3)
```
