```python
#read data
#clean data
#plot graph
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
cig_df=pd.read_excel("C:\Mukul\Data\My Worksheet\Life.xlsx","Cigratte")
cig_df.rename(columns={"Quantity":"Cigratte"},inplace=True)

alcohol_df=pd.read_excel("C:\Mukul\Data\My Worksheet\Life.xlsx","Alcohol")
alcohol_df.rename(columns={"Quantity":"Alcohol"},inplace=True)

weight_df=pd.read_excel("C:\Mukul\Data\My Worksheet\Life.xlsx","Weight")

masturbation_df=pd.read_excel("C:\Mukul\Data\My Worksheet\Life.xlsx","Masturbation")
masturbation_df.rename(columns={"Quantity":"Masturbation"},inplace=True)

gym_df=pd.read_excel("C:\Mukul\Data\My Worksheet\Life.xlsx","Gym")
gym_df.rename(columns={"Quantity":"Gym"},inplace=True)

study_df=pd.read_excel("C:\Mukul\Data\My Worksheet\Life.xlsx","Study")
study_df.rename(columns={"Quantity":"Study"},inplace=True)

diet_df=pd.read_excel("C:\Mukul\Data\My Worksheet\Life.xlsx","Diet")
diet_df.rename(columns={"Quantity":"Diet"},inplace=True)


```

```python
merged_df=pd.concat([cig_df,alcohol_df,weight_df,masturbation_df,gym_df,study_df,diet_df],axis=1)
day=pd.DataFrame(merged_df.iloc[:,1])
date=pd.DataFrame(merged_df.iloc[:,0])
merged_df.drop(["Date","Day"],axis=1,inplace=True)

```

```python
final_df=pd.concat([day[],date,merged_df],axis=1)
```
