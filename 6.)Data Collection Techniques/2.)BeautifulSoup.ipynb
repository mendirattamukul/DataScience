```python
from bs4 import BeautifulSoup
import pandas as pd
books=[]
for i in range(1,51):
    try:
        with open(f"Htmls/page{i}.html","r") as fs:
           soup=BeautifulSoup(fs.read(),"lxml")
           articles=soup.select("article.product_pod")
           for article in articles:
               name=article.h3.a["title"]
               price=article.find("p",class_="price_color").text.split("Â£")[1]
               books.append([name,price])
               
           print(f"Htmls/page{i}.html readed successfully")
    except Exception as e:
        print(e)
df=pd.DataFrame(books,columns=["Name Of Book","Price"])

#.text siraf find ke sath clata hai 
#select mainly css type select karne ke liye use hota hai jaise .select

   
    

```
