```python
import requests as rq
for i in range(1,60):
    with open(f"Htmls/page{i}.html","w",encoding="utf-8") as file:
        response=rq.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
        print(f"https://books.toscrape.com/catalogue/page-{i}.html")
        file.write(response.text) 
        print(f"page{i} Downloaded Successfully") 

    
    
    

            
    

```
