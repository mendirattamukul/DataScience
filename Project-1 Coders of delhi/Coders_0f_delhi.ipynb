```python
import json
def print_details(data):
    print("Users and Their Connections:")
    for user in data["users"]:
        print(f"{user["name"]} (ID: {user["id"]}) - Friends : {user["friends"]} - Liked Pages:{user["liked_pages"]}")
    print("\n\nPages")    
    for page in data["pages"]:  
        print(f"{page["id"]}: {page["name"]}")
        
def load_data(file):
    with open(file,"r") as f:
        data=json.load(f)
    return data
def write_data(data,file_name):
    with open(file_name,"w") as f:
        json.dump(data,f,indent=4)

```

```python
def people_you_may_know(user_id,data):
    suggestion={}
    user_friends={user["id"]:user["friends"] for user in data["users"]}
   
    for friend in user_friends[user_id]:
        for mutual in user_friends[friend]:
            if mutual!=user_id and mutual not in user_friends[user_id]:
                suggestion[mutual]=suggestion.get(mutual,0)+1
    return suggestion
                
                
    
    
   
            
    
```

```python
data=load_data("data2.json")
print(people_you_may_know(1,data))
```

```python
def pages_you_may_like(user_id,data):
    suggestion={}
    user_liked_pages={user["id"]:user["liked_pages"] for user in data["users"]}
    
    
    for user,pages in user_liked_pages.items():
        if user!=user_id:
             shared_pages=(set(user_liked_pages[user_id])).intersection(pages)
             if shared_pages:
                print(user)
                print("\n") 
                pages_that_can_be_liked=set(pages) -set(shared_pages)
                for page in pages_that_can_be_liked:
                    suggestion[page]=suggestion.get(page,0)+len(shared_pages)
    return suggestion
                
        
```
