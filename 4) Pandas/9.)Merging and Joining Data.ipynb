```python
import pandas as pd
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})
 
departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
```

```python
employees
```

```python
departments
```

```python
pd.merge(employees,departments,on="DeptID")#Inner Join
```

```python
pd.merge(employees,departments,on="DeptID",how="left")#left Join
```

```python
pd.merge(employees,departments,on="DeptID",how="right")#right Join
```

```python
pd.merge(employees, departments, on="DeptID", how="outer")#outer join includes all the data and keep missing values as nan
```
