```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [90, 82, 89],
    'English': [88, 85, 94]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
df2=df.melt(id_vars=["Name"],value_vars=["Math","Science"],var_name="Subject",value_name="Score")
```

```python
df2
```

```python
df2.pivot(index="Name",columns=["Subject"],values="Score")
```
