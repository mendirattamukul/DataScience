```python
import numpy as np
arr=np.array([1,2,3,4,5,6])
print(arr.dtype)
```

```python
arr2=np.array([1,5,6.0,7,8,9.2])
print(arr2.dtype)
arr_int=arr2.astype(np.int32)
print(arr_int)
print(arr_int.dtype)
```

```python
arr3=np.array(["apple","mango","banana"],dtype='U10')
print(arr3.dtype)
```

```python
arr4=np.array([1+2j,3j,6j])
arr4.dtype
```

```python
arr5=np.array([1,2,'hello',{'a':1}],dtype="object")
arr5.dtype
```
