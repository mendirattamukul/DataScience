```python
#time comparision for 1 crore record using list
import time


rng=100_000_000
lst1=list(range(rng))
lst2=list(range(rng))

ziplst=zip(lst1,lst2)
list3=list(ziplst)
start_time=time.time()
result=[x+y for x,y in list3]
end_time=time.time()

print(end_time-start_time)
```

```python
#time comparision for 1 crore record using numpy
import time
import numpy as np
rng=100_000_000
lst1=list(range(rng))
lst2=list(range(rng))

lst1=np.array(list(lst1))
lst2=np.array(list(lst2))
start_time=time.time()
result=lst1+lst2
end_time=time.time()

print(end_time-start_time)
```
