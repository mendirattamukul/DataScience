```python
#simple class with static method and class method
class Employee:
    company="HP"
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
        
    def printDetails(self):
        print(f"company ={self.company} name={self.name} salary={self.salary}")
    @staticmethod
    def printTime():
        print("This is static time")
    @classmethod
    def printClassDetails(cls):
        print(cls.company)
        
    
        
e1=Employee("mukul",2600000)
e1.printDetails()
e2=Employee("neha",1600000)
e2.printDetails()
Employee.printTime()
e1.printTime()
e1.printClassDetails()


    
    
        
    
    
```

```python
#Encapsulation 
#preventing direct modification to data instead doing it through getter and settter
class BankAccount:
    def __init__(self,balance):
        self.__balance=balance#private attribute
    def set_balance(self,amount):
        if amount>0:
            self.__balance+=amount
    def get_balance(self):
        return self.__balance
        
account=BankAccount(1000)        
account.set_balance(500)
print(account.get_balance())
```

```python
#single inheritance
class A:
    def print_a(self):
        print("class A")
class B(A):
    def print_b(self):
        print("class B")
obj=B()
obj.print_a()
obj.print_b()

    
    
    
```

```python
#multiple inheritance
class D ():
    def print_d(self):
        print("class d")
class C (A,D):
    def print_c(self):
        print("class c")
obj2=C()
obj2.print_a()
obj2.print_d()
obj2.print_c()
       
```

```python
#Ploymorphism
class Bird:
    def can_fly(self):
        print("Birds can fly")
class Penguin(Bird):
    def can_fly(self):
        print("i cannot fly")
b=Bird()
b.can_fly()
b=Penguin()
b.can_fly()
```

```python
#abstraction
from abc import ABC,abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Square(Shape):
    def __init__(self,side):
        self.side=side
    def area(self):
        return self.side*self.side
square=Square(4)        
print(square.area())
        
        
```

```python
#Dunder Method
class Book:
    def __init__(self,title):
        self.title=title
    def __str__(self):
       return f"Book {self.title}"
    def __len__(self):
        return len(self.title)
b=Book("Python Basics")        
print(str(b))
print(len(b))
```
