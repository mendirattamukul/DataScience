```python
#old Format function
Name="mukul"
Age=28
str="My name is {} and i am {} years old"
print(str.format(Name,Age))
```

```python
#f-string
str=f"My name is {Name} and i am {Age} year old"
print(str)
```

```python
#formating number
pi=3.14567
print(f"rounding up value of pi to 2 decimals {pi:.2f}")
```

```python
#alignment
str="Text"
print(f"{str:<10}")#aligh left
print(f"{str:^10}")#aligh center
print(f"{str:>10}")#aligh right
```

```python
#solving expression inside f-strings
a=10
b=20
print(f"sum of an and b is {a+b}")
```
