##Arithmetic Operators

```python
a=19
b=2
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a**b)
print(a//b)
print(a%b)
```

##Comparision Operators

```python
a=20
b=10
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
print(a==b)
print(a != b)

```

##Logical Operators

```python
a=True
b=False
print(a and b)
print(a or b)
print(not a)
```

##Bitwise Operator

```python
a=17
b=6
print(a&b)
print(a|b)
```
